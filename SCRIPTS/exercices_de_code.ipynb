```python
import yfinance as yf
import pandas as pd 
import numpy as np 
```

```python
ticker = "AAPL"

data = yf.download(ticker, start = "2020-01-01", progress = False, auto_adjust=False)

data.head()  # 5 premières ligne 
data.tail()  # 5 dernière ligne 
```

```python
data["Returns"] = data["Adj Close"].pct_change()

mean_return_daily = data['Returns'].mean()
volatility_daily = data['Returns'].std()

annualized_return = mean_return_daily * 252
annualized_volatility = volatility_daily*np.sqrt(252)
sharpe_ratio = (annualized_return-0.02) / annualized_return

print(f'--- Analyse pour {ticker} ---')
print(f'Rendement Annualisé : {annualized_return}')
print(f'Volatilité Annualisé : {annualized_volatility}')
print(f'Ratio de Sharpe : {sharpe_ratio}')
```
